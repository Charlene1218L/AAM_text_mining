```python
import pandas as pd
```

```python
data=pd.read_csv(r'../Data/interim/all_data_preprocessed_sent_topi.csv')
data
```

topic_change

```python
result = data.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result.to_csv('../Data/output/topic_change_years.csv', encoding="utf_8_sig",index=True)
result
```

topic_change_diffrent_usertype

```python
#The user type is general public
# --  Business Government
data_public=data[data['user_type']=='--']
result = data_public.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../Data/output/topic_general_public.csv', encoding="utf_8_sig",index=True)
result
```

```python
#The user type is Business
#按不同类型用户分 --  Business Government
data_business=data[data['user_type']=='Business']
result = data_business.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../Data/output/topic_business.csv', encoding="utf_8_sig",index=True)
result
```
