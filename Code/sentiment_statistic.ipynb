```python
import pandas as pd
```

```python
data=pd.read_csv(r'../Data/interim/all_data_preprocessed_sent_topi.csv')
data
```

```python
data=data[data['sentiment'].isin(['Negative', 'Neutral', 'Positive'])]
data
```

sentiment_change

```python
result = data.groupby(['year', 'sentiment']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result = result.reset_index()
result.columns.name = None
result.columns = ['year', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
result['sentiment_score']=(result['Negative']*0+result['Neutral']*0.5+result['Positive']*1)/result['Total']
result.to_csv('../Data/output/sentiment_change_years.csv', encoding="utf_8_sig",index=True)
result
```

sentiment_diffirent_usertype

```python
result = data.groupby(['user_type','sentiment']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result.loc['sum_count'] = result.sum()
result = result.reset_index()
result.columns.name = None
result.columns = ['year', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
result['sentiment_score']=(result['Negative']*0+result['Neutral']*0.5+result['Positive']*1)/result['Total']
result.to_csv('../Data/output/sentiment_diffirent_usertype.csv', encoding="utf_8_sig",index=True)
result
```

sentiment_different_nation

```python
result = data.pivot_table(index='nation', columns='sentiment', aggfunc='size', fill_value=0)
result = result.reset_index()
result.columns.name = None
result.columns = ['nation', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
result['sentiment_score']=(result['Negative']*0+result['Neutral']*0.5+result['Positive']*1)/result['Total']
result.to_csv('../Data/output/sentiment_diffirent_nation.csv', encoding="utf_8_sig",index=True)
result
```

Topic_distribution_in_different_sentiment

```python
result = data.groupby(['sentiment', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../Data/output/sentiment_topic.csv', encoding="utf_8_sig",index=True)
result
```
