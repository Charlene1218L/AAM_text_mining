```python
import pandas as pd
```

```python
data=pd.read_csv(r'../../Data/interim/all_data_preprocessed_sent_topi_2.12.csv')
data
```

```python
data=data[data['sentiment'].isin(['Negative', 'Neutral', 'Positive'])]
data
```

sentiment_change

```python
#暂时不用这个
result = data.groupby(['year', 'sentiment']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result = result.reset_index()
result.columns.name = None
result.columns = ['year', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
result['sentiment_score']=(result['Negative']*0+result['Neutral']*0.5+result['Positive']*1)/result['Total']
result.to_csv('../../Data/output/texts/sentiment_change_years_score.csv', encoding="utf_8_sig",index=True)
result
```

sentiment_diffirent_usertype

sentiment_different_nation

```python
result = data.pivot_table(index='nation', columns='sentiment', aggfunc='size', fill_value=0)
result = result.reset_index()
result.columns.name = None
result.columns = ['nation', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
result['sentiment_score']=(result['Negative']*0+result['Neutral']*0.5+result['Positive']*1)/result['Total']
result.to_csv('../../Data/output/texts/sentiment_diffirent_nation_score.csv', encoding="utf_8_sig",index=True)
result
```

ratio_sentiment_change (include positive, neutral, negative)

```python
result = data.groupby(['year', 'sentiment']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result = result.reset_index()
result.columns.name = None
result.columns = ['year', 'Negative', 'Neutral', 'Positive']
result['Total'] = result['Negative'] + result['Neutral'] + result['Positive']
# ratio of positive sentiment posts
result['positive_ratio']=result['Positive']/result['Total']
# ratio of negative sentiment posts
result['negative_ratio']=result['Negative']/result['Total']
# ratio of neutral sentiment posts
result['neutral_ratio']=result['Neutral']/result['Total']
result['max_ratio_sentiment'] = result[['positive_ratio', 'negative_ratio', 'neutral_ratio']].idxmax(axis=1)
result['max_ratio_sentiment'] = result['max_ratio_sentiment'].str[:-6]
result.to_csv('../../Data/output/texts/ratio_sentiment_change_years.csv', encoding="utf_8_sig",index=True)
result
```
