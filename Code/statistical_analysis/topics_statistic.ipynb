```python
import pandas as pd
```

```python
#data=pd.read_csv(r'../Data/interim/all_data_preprocessed_sent_topi.csv')
data=pd.read_csv(r'../../Data/interim/all_data_preprocessed_sent_topi_2.12.csv')
data
```

topic_change

```python
result = data.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum_count'] = result.sum()
result.to_csv('../../Data/output/texts/topic_change_years.csv', encoding="utf_8_sig",index=True)
result
```

topic_change_diffrent_usertype

```python
#The user type is general public
# --  Business Government
data_public=data[data['user_type']=='--']
result = data_public.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../../Data/output/texts/topic_general_public.csv', encoding="utf_8_sig",index=True)
result
```

```python
#The user type is Business
# --  Business Government
data_business=data[data['user_type']=='Business']
result = data_business.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../../Data/output/texts/topic_business.csv', encoding="utf_8_sig",index=True)
result
```

```python
#The user type is Government
# --  Business Government
data_government=data[data['user_type']=='Government']
result = data_government.groupby(['year', 'topic']).size().unstack(fill_value=0)
result.loc['sum'] = result.sum()
result.to_csv('../../Data/output/texts/topic_government.csv', encoding="utf_8_sig",index=True)
result
```

topic_change_diffrent_nation_year

```python
# Count the yearly distribution of each topic for every country
result = data.pivot_table(index=['nation', 'year'], columns='topic', aggfunc='size', fill_value=0)
result.to_csv('../../Data/output/texts/topic_change_diffrent_nation_year.csv', encoding="utf_8_sig",index=True)
result
```
